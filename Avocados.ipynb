```python
#Metodo sin librerias
def makeMatrix(): 
    matrix = []
    file = open("avocado.csv", "r")
    count = 0
    for line in file:
        if count == 0:
            print(line)
        else:
            line = line.strip()
            row = line.split(",")
            matrix.append(row)
        count += 1
    #print("Read ", count, "lines.\n")
    file.close()
    return count
```
